```python
def custom_sorting(df):
    availabiliy = df["Status(Available or not)"]
    for i in range(len(availabiliy)):
        for j in range(0, len(availabiliy) - i - 1):
            if availabiliy[j] > availabiliy[j + 1]:
                temp = df.iloc[j]
                df.iloc[j] = df.iloc[j+1]
                df.iloc[j+1] = temp
```
